2) X Coordinate column

```python
# Step 1: Calculate the mean X Coordinate for each Block
block_mean = df.groupby('Block')['X Coordinate'].transform('mean')

# Step 2: Fill NaN values in X Coordinate with the corresponding block mean
df['X Coordinate'] = df['X Coordinate'].fillna(block_mean)
```

```python
# Step 1: Calculate the mean X Coordinate for '012XX S LAKE SHORE DR NB'
lake_shore_nb_mean = df.loc[df['Block'] == '012XX S LAKE SHORE DR NB', 'X Coordinate'].mean()

# Step 2: Fill NaN values for '012XX S LAKE SHORE DR' using this mean
df.loc[(df['Block'] == '012XX S LAKE SHORE DR') & (df['X Coordinate'].isna()), 'X Coordinate'] = lake_shore_nb_mean
```
